```python
import math
import cv2
import cvzone
from ultralytics import YOLO
```

```python
model= YOLO('./yolo-models/yolov8n.pt')
```

```python
cap = cv2.VideoCapture('./videos/video1.mp4')
mask = cv2.imread('./images/mask.png')
```

```python
cap.set(3,1024)
cap.set(4,576)
while cap.isOpened():
    _ , frame = cap.read()
    frame_height, frame_width, _ = frame.shape
    mask = cv2.resize(mask, (frame_width, frame_height))
    masked_img = cv2.bitwise_and(frame,mask)
    if not _:
        break
    results = model(masked_img,stream=True)
    cv2.line(frame,(73,268),(765,268),(0,0,255),thickness=2)
    for r in results:
        boxes = r.boxes
        for box in boxes:
            
            if box.cls[0] == 2 or box.cls[0] == 7 or box.cls[0] == 3 or box.cls[0] == 5:
                x1,y1,x2,y2 = box.xyxy[0]
                x1,y1,x2,y2 = int(x1) , int(y1) , int(x2) , int(y2)
                w, h = x2-x1 , y2-y1
                cv2.circle()
                cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),thickness=2)
                conf = math.ceil((box.conf[0]*100)) / 100
                cvzone.putTextRect(frame, 'vehicle', (max(0,x1),max(20,y1)),scale=1,offset=6, thickness=1,colorT=(0,0,0))
    cv2.imshow('frame', frame)
    #cv2.setMouseCallback('frame', click_event)
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
cv2.destroyAllWindows()
cap.release()
```
